```python
# Core imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulated dataset (2015-2023)
years = np.arange(2015, 2024)
policy_funding = np.array([2.5, 2.8, 3.1, 3.4, 3.7, 4.0, 4.3, 4.8, 5.2])  # Bn KES (simulated)
irregular_cases = np.array([510, 490, 470, 455, 430, 410, 400, 370, 360])   # (simulated)
trafficking_prosecutions = np.array([15, 18, 22, 25, 27, 30, 34, 37, 40])  # (simulated)
integration_rate = np.array([42, 44, 47, 50, 53, 56, 59, 61, 64])         # percent (simulated)

# Combine into DataFrame
data = pd.DataFrame({
    'Year': years,
    'Policy_Funding_BnKES': policy_funding,
    'Irregular_Cases': irregular_cases,
    'Trafficking_Prosecutions': trafficking_prosecutions,
    'Integration_Rate_%': integration_rate
})

# Summary statistics
print(data.describe())

# Correlation matrix
print(data.corr())

# Plot examples
plt.figure(figsize=(8,5))
plt.plot(data['Year'], data['Policy_Funding_BnKES'], label='Policy Funding (Bn KES)')
plt.plot(data['Year'], data['Integration_Rate_%'], label='Integration Rate (%)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Policy Funding vs. Refugee Integration Rate in Kenya (2015â€“2023)')
plt.legend()
plt.show()
```
